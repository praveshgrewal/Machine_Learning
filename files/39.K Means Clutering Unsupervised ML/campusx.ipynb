```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("student_clustering.csv")
df.shape
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11))
```

```python

x= df.iloc[:,:].values

km = KMeans(n_clusters=4)
y_mean= km.fit_predict(x)
```

```python
y_mean
```

```python
x[y_mean == 2,1]
```

```python
plt.scatter(x[y_mean == 0,0],x[y_mean == 0,1],color = 'blue')
plt.scatter(x[y_mean == 1,0],x[y_mean == 1,1],color = 'red')
plt.scatter(x[y_mean == 2,0],x[y_mean == 2,1],color = 'yellow')
plt.scatter(x[y_mean == 3,0],x[y_mean == 3,1],color = 'green')
```

# 3D

```python
from sklearn.datasets import make_blobs
```

```python
cent = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
clu= [1,1,1,1]
```

```python
x,y= make_blobs(n_samples=200,cluster_std = clu, centers=cent,n_features=3,random_state=42)
```

```python
x
```

```python
import plotly.express as px
fig = px.scatter_3d(x= x[:,0],y= x[:,1],z=x[:,2])
fig.show()
```

```python
wcss=[]

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
plt.plot (range(1,11),wcss)
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(x)
```

```python
df = pd.DataFrame()

df['col1'] = x[:,0]
df['col2'] = x[:,1]
df['col3'] = x[:,2]
df['label'] = y_pred
```

```python
fig = px.scatter_3d(df,x='col1',y= 'col2',z='col3',color='label')
fig.show()
```
